```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("ccdefault.csv", index_col="ID")
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt

X, y = df.drop(["DEFAULT"], axis=1), df['DEFAULT']

train_accuracy = []
test_accuracy = []

for i in range(1, 11):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=i, stratify=y)
    
    dtc = DecisionTreeClassifier(max_depth=5, min_samples_leaf=4, min_samples_split=2, random_state=42)
    dtc = dtc.fit(X_train, y_train)
    
    y_pred_train = dtc.predict(X_train)
    y_pred = dtc.predict(X_test)
    
    train_accuracy.append(accuracy_score(y_train, y_pred_train))
    test_accuracy.append(accuracy_score(y_test, y_pred))
    
df_score = pd.DataFrame({"Random State": np.arange(1, 11), "Train": train_accuracy, "Test": test_accuracy})
df_score.set_index('Random State', inplace=True)
df_score.T
```

```python
df_stat = { "Train": [np.mean(train_accuracy), np.std(train_accuracy)],"Test": [np.mean(test_accuracy), np.std(test_accuracy)]}
df_stat = pd.DataFrame(df_stat, index=['Mean', 'SD'])
df_stat.T
```

```python
from sklearn.model_selection import StratifiedKFold, cross_validate

dtc = DecisionTreeClassifier(max_depth=5, min_samples_leaf=4, min_samples_split=2, random_state=42)
skf = StratifiedKFold(n_splits=10)

scores = cross_validate(dtc, X, y, cv=skf)
df_score_K = pd.DataFrame(
    {'K': np.arange(1, 11), 'Test': scores['test_score']})
df_score_K.set_index('K', inplace=True)
df_score_K.T
```

```python
df_stat_K = { "Stat": [np.mean(scores['test_score']), np.std(scores['test_score'])]}
df_stat_K = pd.DataFrame(df_stat_K, index=['Mean', 'SD'])
df_stat_K.T
```
